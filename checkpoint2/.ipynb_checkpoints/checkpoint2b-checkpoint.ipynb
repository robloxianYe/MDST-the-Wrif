```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# feel free to import other libraries! 
```

```python
df = pd.read_csv('../data/states_edu.csv')
```

Chosen test: Grade 8 Math

<h2> Cleanup (optional)</h2>

_Use this space to rename columns, deal with missing data, etc._

```python
df.dropna(subset = ["ENROLL"], inplace=True)
df.dropna(subset = ["AVG_MATH_8_SCORE"], inplace=True)
df.dropna(subset = ["INSTRUCTION_EXPENDITURE"], inplace=True)
df.dropna(subset = ["SUPPORT_SERVICES_EXPENDITURE"], inplace=True)
```

<h2> Feature Selection </h2>

_Use this space to modify or create features_

```python
df["SUPPORT_SERVICES_EXPENDITURE_PER_STUDENT"] = df["SUPPORT_SERVICES_EXPENDITURE"] / df["ENROLL"]
df["INSTRUCTION_EXPENDITURE_PER_STUDENT"] = df["INSTRUCTION_EXPENDITURE"] / df["ENROLL"]
```

Final feature list: Data representing the support services expenditure per student and data representing instruction expenditure per student

Feature selection justification: Both instruction expenditure and support service expenditure could account for differences in test scores. However, states with higher enrollment will obviously spend more. Therefore, a measure of expenditure per student is a better way to analyze the data.

<h2> EDA </h2>

Visualization 1

```python
df.AVG_MATH_8_SCORE.hist()
plt.xlabel('score')
plt.ylabel('freq')
plt.title('Distribution of 8th grade math scores')
```

The distribution is unimodal skewed left centered around about 284/500 (Math scores are out of 500). The maximum is about 300 while the minimum is at around 235.

Visualization 2

```python
df.plot.scatter(x='INSTRUCTION_EXPENDITURE_PER_STUDENT', y='AVG_MATH_8_SCORE')
plt.xlabel('State instruction expenditure per student')
plt.ylabel('State average 8th grade math score')
plt.title('8th grade math score compared with instruction expenditures')
```

There does appear to be a noticeable positive correlation between instruction expenditure and math scores for 8th graders

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[["SUPPORT_SERVICES_EXPENDITURE_PER_STUDENT", "INSTRUCTION_EXPENDITURE_PER_STUDENT", "AVG_MATH_8_SCORE"]]
y = df.loc[X.index]["AVG_MATH_8_SCORE"]
```

```python
df.AVG_MATH_4_SCORE.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.3, random_state=0)
```

<h2> Prediction </h2>

Chosen ML task: Regression

```python
# import your sklearn class here
from sklearn.linear_model import LinearRegression
```

```python
# create your model here
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_train
```

```python
y_pred = model.predict(X_test)
print(model.intercept_)
print(model.coef_)
print(model.score(X_test, y_test))
print(np.mean(np.abs(model.predict(X_test)-y_test)))
```

```python
# FOR CLASSIFICATION ONLY:
#from sklearn.metrics import plot_confusion_matrix

#plot_confusion_matrix(model, X_test, y_test,
#                         cmap=plt.cm.Blues)
```

```python
# FOR REGRESSION ONLY: (pick a single column to visualize results)

# Results from this graph _should not_ be used as a part of your results -- it is just here to help with intuition. 
# Instead, look at the error values and individual intercepts.


col_name = "INSTRUCTION_EXPENDITURE_PER_STUDENT"
col_index = X_train.columns.get_loc(col_name)

f = plt.figure(figsize=(12,6))
plt.scatter(X_train[col_name], y_train, color = "red")
plt.scatter(X_train[col_name], model.predict(X_train), color = "green")
plt.scatter(X_test[col_name], model.predict(X_test), color = "blue")

new_x = np.linspace(X_train[col_name].min(),X_train[col_name].max(),100)
intercept = model.predict([X_train.sort_values(col_name).iloc[0]]) - X_train[col_name].min()*model.coef_[col_index]
plt.plot(new_x, intercept+new_x*model.coef_[col_index])

plt.legend(['controlled model','true training','predicted training','predicted testing'])
plt.xlabel(col_name)
plt.ylabel("Math Grade 8 score")
```

<h2> Summary </h2>

There is a positive correlation between instruction expenditure and average math scores as well as support service expenditure and average math grades. The correlation between instruction spending and grades seems to be slightly better than the correlation between support service spending and grades.
